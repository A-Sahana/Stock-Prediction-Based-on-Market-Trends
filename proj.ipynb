```python
import numpy as np

# Create sample data
x_train = np.random.rand(1000, 10, 1)  #1000 samples 10 time step 1 feature
y_train = np.random.rand(1000, 1)  #1000 samples 1 target variable
x_test = np.random.rand(200, 10, 1)   #200 samples 10 time steps 1 feature
y_test = np.random.rand(200, 1)  #200 samples 1 target variable

#print statements that show thw shape of training and testing
print(f'x_train shape: {x_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'x_test shape: {x_test.shape}')
print(f'y_test shape: {y_test.shape}')

#define time_step and number_of_features based on x_train
time_step = x_train.shape[1]  #2nd dimension of x_train
number_of_features = x_train.shape[2]
print(f'time_step: {time_step}, number_of_features: {number_of_features}')
```

```python
import numpy as np
import math
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Assuming your dataset is already loaded into the variables x_train, y_train, x_test, and y_test

print(f'x_train shape: {x_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'x_test shape: {x_test.shape}')
print(f'y_test shape: {y_test.shape}')

# Reshaping y_train and y_test if they are 1D arrays
if y_train.ndim == 1:
    y_train = np.reshape(y_train, (len(y_train), 1))
if y_test.ndim == 1:
    y_test = np.reshape(y_test, (len(y_test), 1))

print(f'y_train shape after reshape: {y_train.shape}')
print(f'y_test shape after reshape: {y_test.shape}')

# Define time_step and number_of_features based on x_train
if len(x_train.shape) == 3:
    time_step = x_train.shape[1]
    number_of_features = x_train.shape[2]
    print(f'time_step: {time_step}, number_of_features: {number_of_features}')
else:
    raise ValueError("x_train does not have the expected 3D shape.")

# Creating the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, number_of_features)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Initialize and fit scaler for the target variable
scaler = MinMaxScaler(feature_range=(0, 1))
scaler = scaler.fit(y_train)  # Fit scaler on y_train data

# Transform y_train and y_test
y_train = scaler.transform(y_train)
y_test = scaler.transform(y_test)

# Fit the model
model.fit(x_train, y_train, batch_size=32, epochs=10)  # Adjust batch_size and epochs as needed

# Predict
train_predict = model.predict(x_train)
test_predict = model.predict(x_test)

# Inverse transform predictions and actual values
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train = scaler.inverse_transform(y_train)
y_test = scaler.inverse_transform(y_test)

# Evaluate the model
train_rmse = math.sqrt(mean_squared_error(y_train, train_predict))
test_rmse = math.sqrt(mean_squared_error(y_test, test_predict))

print(f'Train RMSE: {train_rmse}')
print(f'Test RMSE: {test_rmse}')
```

```python
model.save('stock_model.h5')
import joblib
joblib.dump(scaler, 'scaler.pkl')
```
